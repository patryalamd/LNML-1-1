```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('mba_decision_dataset.csv')
df.head()  # Tampilkan beberapa baris pertama dataset

```

```python
# Cek informasi umum tentang dataset
df.info()

# Cek statistik deskriptif
df.describe()

# Cek missing values
df.isnull().sum()

```

```python
# Visualisasi distribusi usia
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], kde=True)
plt.title('Distribusi Usia')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.show()

```

```python
# Visualisasi hubungan antara Usia dan Keputusan untuk Mengejar MBA
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Age', y='Decided to Pursue MBA?', hue='Decided to Pursue MBA?')
plt.title('Hubungan antara Usia dan Keputusan untuk Mengejar MBA')
plt.xlabel('Usia')
plt.ylabel('Keputusan untuk Mengejar MBA')
plt.show()
```

```python
# Pilih hanya kolom numerik
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Hitung korelasi antar fitur numerik
corr = df[numeric_columns].corr()

# Visualisasi korelasi menggunakan heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Korelasi antar Fitur Numerik')
plt.show()

```

```python
# Visualisasi distribusi Gender berdasarkan keputusan untuk mengejar MBA
plt.figure(figsize=(10,6))
sns.countplot(x='Gender', hue='Decided to Pursue MBA?', data=df)
plt.title('Distribusi Gender Berdasarkan Keputusan Mengejar MBA')
plt.xlabel('Gender')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Visualisasi distribusi keputusan untuk mengejar MBA
plt.figure(figsize=(8,6))
sns.countplot(x='Decided to Pursue MBA?', data=df)
plt.title('Distribusi Keputusan untuk Mengejar MBA')
plt.xlabel('Keputusan untuk Mengejar MBA')
plt.ylabel('Frekuensi')
plt.show()
```

Insight 1: Distribusi Usia
"Dari visualisasi distribusi usia, kita melihat bahwa sebagian besar responden berada dalam rentang usia 25-35 tahun. Ini menunjukkan bahwa banyak orang memutuskan untuk mengejar MBA setelah beberapa tahun pengalaman kerja setelah menyelesaikan sarjana."

Insight 2: Hubungan Usia dan Keputusan untuk Mengejar MBA
"Scatter plot ini menunjukkan hubungan antara usia dan keputusan untuk mengejar MBA. Kita bisa melihat bahwa individu yang lebih muda lebih cenderung memilih untuk mengejar MBA, mungkin karena mereka memiliki lebih sedikit kewajiban dan lebih fleksibel dalam memilih jalur pendidikan lebih lanjut."

Insight 3: Korelasi Antar Fitur Numerik
"Heatmap korelasi mengungkapkan bahwa terdapat korelasi yang cukup tinggi antara GPA Sarjana dan Skor GRE/GMAT, yang mengindikasikan bahwa individu dengan prestasi akademik yang baik di tingkat sarjana cenderung juga memiliki skor ujian yang tinggi. Ini dapat membantu dalam menentukan faktor-faktor yang mempengaruhi keputusan untuk melanjutkan pendidikan."

Insight 4: Distribusi Gender Berdasarkan Keputusan untuk Mengejar MBA
"Distribusi gender berdasarkan keputusan untuk mengejar MBA menunjukkan bahwa meskipun lebih banyak pria yang memilih MBA, tetapi perbedaan antara gender tidak terlalu signifikan, yang mengindikasikan bahwa program MBA cukup inklusif bagi berbagai gender."

Insight 5: Distribusi Keputusan untuk Mengejar MBA
"Dari count plot ini, terlihat bahwa mayoritas responden memutuskan untuk mengejar MBA, yang menunjukkan bahwa program MBA sangat diminati dan dianggap sebagai jalur penting untuk perkembangan karir."
